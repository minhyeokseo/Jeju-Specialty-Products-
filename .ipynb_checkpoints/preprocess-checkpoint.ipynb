# Preprocessing Selling Data

## Load Data

```python
selling=pd.read_csv('./data/train.csv')
selling
```

## Change Dateformat & ID

```python
for idx in range(len(selling)):
    selling.loc[idx,'ID']=selling.loc[idx,'ID'][:-2]
    selling.loc[idx,'timestamp']=selling.loc[idx,'timestamp'][:-3]
selling
```

## Grouping with Timestamp

```python
mon_selling=selling.iloc[:,[0,1,2,5,6]].groupby(['ID','timestamp','item']).sum()
mon_selling
```

```python
fig = plt.figure(figsize=(20, 8))
sns.lineplot(data=mon_selling,
             x='timestamp',
             y='supply(kg)',
             hue='item')
plt.xticks(rotation=45)
plt.show()
```

# ASOS Data

```python
url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'
date=['20190101','20191231','20200101','20201231','20210101','20211231','20220101','20221231','20230101','20230304']
row=['365','366','365','365','62']


asos_df=0
for stn in [184,189]:
    df1=0
    for i in range(0,len(date),2):
        params ={'serviceKey' : 'wACC0jmLBit52BZepbj7OeworqNhenCQapfqcO/wJO86y23ikIAIRESSY1u9YOm/PJXYn/D9hQQElLshrtn8zQ==',
                 'pageNo' : '1',
                 'numOfRows' : row[i//2],
                 'dataType' : 'json',
                 'dataCd' : 'ASOS',
                 'dateCd' : 'DAY',
                 'startDt' : date[i],
                 'endDt' : date[i+1],
                 'stnIds' : stn }

        response = requests.get(url, params=params)
        json_ob = json.loads(response.content)
        json_ob=json_ob['response']['body']['items']['item']
        df=pd.DataFrame(json_ob)
        
        if type(df1)!=int:
            df1=pd.concat([df1,df]).reset_index(drop=True)
        else:
            df1=df.copy()
    
    if type(asos_df)!=int:
        asos_df=pd.concat([asos_df,df1]).reset_index(drop=True)
    else:
        asos_df=df1.copy()

asos_df
```
